데이터 분석 과정에서 데이터 전처리가 중요한 근본적 이유는 아무리 좋은 분석 기법이나 도구를 사용한다 하더라도 품질이 낮은 데이터로는 좋은 분석 결과를 얻을 수 없기 때문이다. 데이터 품질을 떨어뜨리는 주요 원인은 다음과 같다.  

1. 불완전(incomplete)    
- 데이터가 비어 있는 경우로 DB 테이블 속성 값이 NULL인 경우  
2. 잡음(noisy)  
- 데이터에 오류가 포함된 경우 예) 나이 = -20  
3. 모순된(inconsistent)  
- 데이터 간의 정합성(일관성)이 없는 경우. 예) 성별은 남자인데, 주민번호 뒤 7자리 중 첫 번째 자리가 2인 경우  

낮은 데이터 품질이 데이터 전처리를 필요로 하는 가장 중요한 이유이기는 하지만, 데이터 품질이 높다 하더라도 데이터 전처리를 필요로 하는 경우도 많다. 실무에서 존재하는 데이터의 구조적인 형태가 분석 목적이나 분석 기법에 적합한 경우가 드물기 때문이다. 데이터의 구조적 형태를 분석 기법을 적용하기 용이한 형태로 전환시키는 것도 데이터 전처리 영역에 속한다.

판다스(Pandas)는 파이썬에서 데이터 처리를 위해 존재하는 가장 인기있는 라이브러리입니다. 일반적으로 대부분의 데이터 세트는 2차원 데이터입니다. 즉, 행(Row) x 열(Column)으로 구성되어있습니다. 행과 열의 2차원 데이터가 인기 있는 이유는 바로 인간이 가장 이해하기 쉬운 데이터 구조이면서도 효과적으로 데이터를 담을 수 있는 구조이기 때문입니다. 판다스는 이처럼 행과 열로 이뤄진 2차원 데이터를 효율적으로 가공/처리할 수 있는 다양하고 훌륭한 기능을 제공합니다.  

앞 절에서 넘파이를 소개했지만, 넘파이의 데이터 핸들링은 편하다고 말하기 어렵습니다. 넘파이는 저수준 API가 대부분입니다. 판다스는 많은 부분 넘파이 기반으로 작성됐는데, 넘파이보다 훨씬 유연하고 편리하게 데이터 핸들링을 가능하게 해줍니다. 판다스는 파이썬의 리스트, 컬렉션, 넘파이 등의 내부 데이터뿐만 아니라 CSV 등의 파일을 쉽게 DataFrame으로 변경해 데이터의 가공/분석을 편리하게 수행할 수 있게 만들어줍니다.  
판다스가 파이썬 세계의 대표적인 데이터 핸들링 프레임워크지만, 광범위한 영역을 커버하고 있기에 이를 마스터하기에는 의외로 많은 노력과 시간 투자가 필요합니다. 판다스 기능 자체에 대한 설명만 하더라도 별도의 책 한권 분량의 내용이 될 정도입니다.  

판다스의 핵심 객체는 DataFrame입니다. DataFrame은 여러 개의 행과 열로 이뤄진 2차원 데이터를 담는 데이터 구조체입니다. 판다스가 다루는 대부분의 영역은 바로 DataFrame에 관련된 부분입니다. DataFrame을 이해하기 전에 다른 중요 객체인 Index와 Series를 이해하는 것도 중요합니다. Index는 RDBMS의 PK처럼 개별 데이터를 고유하게 식별하는 Key 값입니다. Series와 DataFrame은 모두 Index를 키 값으로 가지고 있습니다. 시리즈와 데이터프레임의 가장 큰 차이는 시리즈는 칼럼이 하나 뿐인 데이터 구조체이고, 데이터프레임은 칼럼이 여러 개인 데이터 구조체라는 점입니다.  

데이터프레임은 여러 개의 시리즈로 이뤄졌다고 할 수 있습니다. 데이터프레임을 익히기 위해서는 인덱스와 시리즈를 먼저 이해하는 것이 필요하지만 데이터프레임을 개괄적인 수준에서 이해한 후에 인덱스와 시리즈를 이해하는 것이 더 쉽고 편리한 방법일 수도 있습니다.

판다스는 결손 데이터(Missing Data)를 처리하는 편리한 API를 제공합니다. 결손 데이터는 칼럼에 값이 없는, 즉 NULL인 경우를 의미하며, 이를 넘파이의 NaN으로 표시합니다. 기본적으로 머신러닝 알고리즘은 이 NaN 값을 처리하지 않으므로 이 값을 다른 값으로 대체해야 합니다. 또한 NaN 값은 평균, 총합 등의 함수 연산 시 제외가 됩니다. 특정 칼럼의 100개 데이터 중 10개가 NaN 값일 경우, 이 칼럼의 평균 값은 90개 데이터에 대한 평균입니다. p. 38 특징값 추출

특징값 추출(feature extraction) 단계는 데이터 전처리 및 탐험 단계와 매우 밀접합니다. 특징값은 말 그대로 데이터의 여러 특징을 나타내는 값입니다. 문자열 데이터를 예로 들면 문서에 있는 단어의 평균 길이나 단어 수 등이 문서의 특징값이 될 수 있습니다. 다양한 지역을 다루는 데이터라면 각 지역의 월 평균 온도가 각 지역을 나타내는 특징값이 될 수 있죠.

특징값을 추출하는 과정은 날것 그대로의 데이터를 가공해서 그 값을 표에 열심히 저장하는 단순한 일입니다. 이렇게 표에 저장된 데이터를 표 데이터(tabular data)라고 합니다. 보통 각 행이 각각 데이터 표본 하나를 의미하고 열마다 다른 종류의 데이터를 저장합니다. 예를 들어 여러 문서로 이루어진 데이터라면 행마다 다른 문서를 나타내는 특징값을 저장합니다. 따라서 전체 행수는 문서 수가 되고, 전체 열 수는 특징값 수가 됩니다. 대부분의 데이터 분석 방법은 표 데이터를 가지고 시작합니다.

특징값 추출은 뒤에서 배울 머신러닝 분류기나 멋진 최신 통계 모델, 아름답고 깔끔한 코드 구현보다 훨씬 중요합니다. 어쩌면 데이터 분석 단계 중 가장 중요합니다. 특히 데이터가 웹페이지나 이미지인 경우 특징값을 반드시 추출해야 하는데, 효과적인 특징값 추출이 전체 데이터 분석의 성패를 좌우합니다.  

특징값 추출은 데이터 과학에서 가장 창의성을 발휘할 수 있는 작업입니다. 반면 각 데이터의 특성을 알아야 하기 때문에 여러 분야에 걸친 배경지식이 필요합니다. 예를 들어 문서, 웹 페이지, 음성, 사진은 각각 다른 특성을 가지고 있기 때문에 특징값의 의미와 추출 방법이 각기 다릅니다. 좋은 특징값은 우리가 해석 가능한 어떤 의미를 나타냅니다. 그리고 데이터의 배경지식을 가지고 있어야 이런 의미가 무엇이고 그 의미에 해당하는 값을 뽑아내려면 어떻게 해야 하는지 알 수 있습니다.  

데이터 종류마다 널리 알려진 특징값이 있지만 때론 스스로 특징값을 만들어야 하는 경우도 있습니다. 예를 들어 각 거래 기록마다 해당하는 사람의 이름과 이메일 주소가 적힌 상거래 데이터를 가정합시다. 만일 한 사람이 이메일 주소를 하나만 사용한다고 가정하면 전체 이메일 수로 거래에 관여한 사람 수를 셀 수 있습니다. 반면 이름 데이터는 겹칠 수 있으므로 동명이인을 감안해서 아뤄야 합니다.  

이렇게 뽑은 특징값은 나중에 뭔가를 예측하는 데 쓰이게 될 겁니다. 때론 이 예측값이 데이터에 없는 경우도 있고 데이터 과학자가 직접 만들어야 하는 경우도 있습니다.


p.39 모델 수집 및 분석

특징값을 뽑고 나면 이제 본격적으로 머신러닝 모델을 사용할 차례입니다. 고객의 충성도를 0에서 1 사이의 값으로 나타내기로 했다면 회귀(regression) 모델을 사용하고, 다양한 고객을 몇 가지 범주로 분류하고자 한다면 군집화(clustering) 알고리즘을 사용합니다.  

대개 모델을 정하는 과정은 그리 복잡하지 않습니다. 흔희 쓰는 표준화된 모델 중 적절한 것을 몇 개 고르고, 데이터를 넣어서 성능을 비교하면 됩니다. 다만 모델을 처음부터 새로 만들어야하는 경우도 드물게 있습니다.  

그러나 단순히 모델을 돌려보는 것 말고도 왜 모델이 이렇게 작동하는지 질문을 던져볼 필요가 있습니다. 예를 들어 군집화를 했다면 각 군집이 어떤 특징을 가지는지, 분류기 성능이 만족스러운 수준인지, 분류가 잘 안 되는 데이터 표본은 어떤 특징을 가지고 있는지 등이 있습니다. 이 과정을 통해 문제점을 파악해 모델을 개선할 수 있습니다.  

만일 사람에게 결과를 설명해야 한다면 여러 모델을 사용해보고 그 장단점을 보고하는 경우가 많습니다. 반면 컴퓨터가 고객이라면, 즉 코드를 배포해야 한다면 최종적으로 한 모델을 선정하게 됩니다.

### 4. 데이터 정제하기


```python
# 앞의 값으로 빈 값 채우기
df2 = df.fillna(method='bfill')
```

```python
# 뒤의 값으로 빈 값 채우기
df.fillna(method='ffill')
```

```python
# 빈 값 없애기
df2=  df.dropna()
```

```python
# 앞의 5개 데이터만 살펴보기
df2.head()
```

```python
# 데이터 저장하기
import numpy as np
X = np.array(df2['년'])
y = np.array(df2['평균기온(℃)'])
type(X)

print(X.shape, y.shape)
```

```python
# 훈련셋, 테스트셋 나누기
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y,shuffle=False)
print(X_train.shape, X_test.shape, y_train.shape, y_test.shape)
X_train = np.reshape(X_train,(-1,1))
X_test = np.reshape(X_test,(-1,1))
print(X_train.shape, X_test.shape, y_train.shape, y_test.shape)
```

선형 모델은 100여 년 전에 개발되었고, 지난 몇십 년 동안 폭넓게 연구되고 현재도 널리 쓰입니다. 곧 보겠지만 선형 모델은 입력 특성에 대한 선형 함수를 만들어 예측을 수행합니다.

```python
# 알고리즘 선택 - 모델 만들기
from sklearn.linear_model import LinearRegression
model = LinearRegression().fit(X_train, y_train)
```

```python
# 예측하기
y_pred = model.predict(X_test)
y_pred
```

```python
# 차이 비교하기
print(np.mean(y_test-y_pred))
```
