```python
import pandas as pd
import matplotlib.pyplot as plt
plt.figure(dpi=300)
plt.rc('font', family = 'Malgun Gothic')
plt.title('서울의 기온 변화')
df = pd.read_csv('seoul_temp.csv', encoding = 'cp949')
plt.plot(df['최저기온'],'skyblue', label = '최저기온')
plt.plot(df['최고기온'],'hotpink', label = '최고기온')
plt.legend()
plt.show()
```
